# 1.Preprocessing Data

```python
import os

# Define the relative path from the script's directory
relative_folder_path = "Desktop/collecting data/week6A4/17 Modernist British and American Poetry"

# Construct the absolute folder path
folder_path = os.path.join(os.path.expanduser("~"), relative_folder_path)

file_contents = []

# Read file contents one by one
for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    if os.path.isfile(file_path) and file_name.endswith('.txt'):
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            file_contents.append(content)

# Print the first file and check and compare it with the original file
print(file_contents[0])
```

①Create DataFrame and extract metadata information from files using the Pandas library in Python

```python
import re
import pandas as pd
folder_path = os.path.join(os.path.expanduser("~"), relative_folder_path)

# Sort by file name (1-17)
file_names = sorted(os.listdir(folder_path), key=lambda x: int(re.search(r'(\d+)', x).group()))

# Create an empty DataFrame to store file information
data = pd.DataFrame(columns=['Filename', 'Title', 'Release Date', 'Author'])

# Read file contents one by one and extract information
for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)
    if os.path.isfile(file_path) and file_name.endswith('.txt'):
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            
            title_match = re.search(r'Title:\s*(.*?)\s*Author:', content, re.DOTALL)
            release_date_match = re.search(r'Release Date:\s*(.*?)\s*\[', content, re.DOTALL)
            author_match = re.search(r'Author:\s*(.*?)\s*Release Date:', content, re.DOTALL)
            
            ##Correctly obtain matching content
            title = title_match.group(1).strip() if title_match else ''
            release_date = release_date_match.group(1).strip() if release_date_match else ''
            author = author_match.group(1).strip() if author_match else ''
            
            data = data.append({
                'Filename': file_name,
                'Title': title,
                'Release Date': release_date,
                'Author': author,
            }, ignore_index=True)

##Show extracted content
data
```

# 2.Data Cleansing

```python
import os
import re

folder_path = os.path.join(os.path.expanduser("~"), relative_folder_path)
file_names = os.listdir(folder_path)
file_names_sorted = sorted(file_names, key=lambda x: int(re.search(r'\d+', x).group()))

file_contents = []

# Function to clean text from HTML tags 
def clean_text(text):
    # Remove HTML tags using regex
    clean_text = re.sub('<.*?>', '', text)
    return clean_text

# Read file contents one by one and clean the text
for file_name in file_names_sorted:
    file_path = os.path.join(folder_path, file_name)
    if os.path.isfile(file_path) and file_name.endswith('.txt'):
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            cleaned_content = clean_text(content)
            file_contents.append(cleaned_content)


print(file_contents[5])
```

```python
import spacy

# Install English language model
nlp = spacy.load("en_core_web_sm")
folder_path = os.path.join(os.path.expanduser("~"), relative_folder_path)
tokenized_texts = []

##Sort by file name (1-17)
file_names = sorted((f for f in os.listdir(folder_path) if f.endswith('.txt')), key=lambda x: int(''.join(filter(str.isdigit, x))))
##Read each file in order and tokenize it
for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        #Using spacy to tokenize content
        doc = nlp(content)
        
        tokenized_text = [token.text for token in doc]
        tokenized_texts.append(tokenized_text)
        
print(tokenized_texts[13])
```

```python
folder_path = os.path.join(os.path.expanduser("~"), relative_folder_path)
lemmatized_texts = []

#Sort by file name (1-17)
file_names = sorted((f for f in os.listdir(folder_path) if f.endswith('.txt')), key=lambda x: int(''.join(filter(str.isdigit, x))))

#Read the file content one by one and merge the word forms
for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        
        doc = nlp(content)
        
        lemmatized_text = [token.lemma_ for token in doc]
        lemmatized_texts.append(lemmatized_text)


print(lemmatized_texts[0])
```

```python
folder_path = os.path.join(os.path.expanduser("~"), relative_folder_path)
pos_tags_list = []

##Sort by file name (1-17)
file_names = sorted((f for f in os.listdir(folder_path) if f.endswith('.txt')), key=lambda x: int(''.join(filter(str.isdigit, x))))

##Read file contents one by one and annotate part of speech
for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        
        doc = nlp(content)
       
        pos_tags = [token.pos_ for token in doc]
        pos_tags_list.append(pos_tags)

print(pos_tags_list[8])
```

```python
df = pd.DataFrame(data, columns=['Filename', 'Title', 'Release Date', 'Author'])
df['Tokenize'] = tokenized_texts  
df['POS_tags'] = pos_tags_list  
df['Lemmatized'] = lemmatized_texts 

# Save the DataFrame to a CSV file
df.to_csv('Modernist_British_and_American_Poetry.csv', index=False)
```

An extra column with the original (untokenized) text of each poem.

```python
import os
import pandas as pd

# Folder path containing text files
folder_path = os.path.join(os.path.expanduser("~"), relative_folder_path)

# CSV file path
csv_file_path = r"C:\Users\yd180\Desktop\collecting data\week6A4\Modernist_British_and_American_Poetry.csv"

# Read existing CSV file into a DataFrame
data = pd.read_csv(csv_file_path, encoding='utf-8')

# Function to extract text from a file and remove line breaks
def extract_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read().replace('\n', ' ')
    return text

# Add a new column to store extracted text
data['Text'] = ''

# Update the 'Text' column with the extracted text
for index, row in data.iterrows():
    file_name = row['Filename']
    file_path = os.path.join(folder_path, file_name)
    if os.path.isfile(file_path) and file_name.endswith('.txt'):
        text = extract_text(file_path)
        data.at[index, 'Text'] = text

# Save the updated DataFrame to the CSV file with utf-8 encoding
data.to_csv(csv_file_path, index=False, encoding='utf-8')

print("Text extraction and update completed.")
```
